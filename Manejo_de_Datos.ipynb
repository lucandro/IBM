Cada lanzamiento tiene como objetivo una órbita dedicada, y estos son algunos tipos de órbitas comunes:

- LEO: la órbita terrestre baja (LEO) es una órbita centrada en la Tierra con una altitud de 2000 km (1200 mi) o menos (aproximadamente un tercio del radio de la Tierra),[1] o con al menos 11,25 períodos por día (un período orbital de 128 minutos o menos) y una excentricidad menor a 0,25.[2] La mayoría de los objetos creados por el hombre en el espacio exterior están en LEO [1].

- VLEO: las órbitas terrestres muy bajas (VLEO) se pueden definir como las órbitas con una altitud media por debajo de los 450 km. Operar en estas órbitas puede proporcionar una serie de beneficios a las naves espaciales de observación de la Tierra, ya que la nave espacial opera más cerca de la observación[2].

- GTO Una órbita geosincrónica es una órbita terrestre alta que permite a los satélites coincidir con la rotación de la Tierra. Ubicado a 22,236 millas (35,786 kilómetros) sobre el ecuador de la Tierra, esta posición es un punto valioso para monitorear el clima, las comunicaciones y la vigilancia. Debido a que el satélite orbita a la misma velocidad que gira la Tierra, el satélite parece permanecer en su lugar a lo largo de una única longitud, aunque puede desplazarse de norte a sur”, escribió la NASA en su sitio web Earth Observatory [3] .

- SSO (o SO): es una órbita heliosincrónica, también llamada órbita heliosincrónica, es una órbita casi polar alrededor de un planeta, en la que el satélite pasa sobre cualquier punto dado de la superficie del planeta a la misma hora solar media local [4] .

- ES-L1: en los puntos de Lagrange, las fuerzas gravitacionales de los dos cuerpos grandes se cancelan de tal manera que un objeto pequeño colocado en órbita allí está en equilibrio en relación con el centro de masa de los cuerpos grandes. L1 es uno de esos puntos entre el sol y la tierra [5] .

- HEO Órbita altamente elíptica, es una órbita elíptica con alta excentricidad, generalmente referida a una órbita alrededor de la Tierra [6].

- ISS Estación espacial modular (satélite artificial habitable) en órbita terrestre baja. Es un proyecto colaborativo multinacional entre cinco agencias espaciales participantes: NASA (Estados Unidos), Roscosmos (Rusia), JAXA (Japón), ESA (Europa) y CSA (Canadá) [7]

- MEO Órbitas geocéntricas que varían en altitud desde 2000 km (1200 mi) hasta justo por debajo de la órbita geosincrónica a 35 786 kilómetros (22 236 mi). También conocida como órbita circular intermedia. Estos se encuentran "más comúnmente a 20.200 kilómetros (12.600 mi), o 20.650 kilómetros (12.830 mi), con un período orbital de 12 horas [8]

- HEO Órbitas geocéntricas por encima de la altitud de la órbita geosincrónica (35.786 km o 22.236 mi) [9]

```python
# landing_outcomes = valores en la columna Resultado
landing_outcomes = df["Outcome"].value_counts()
landing_outcomes
```

`True Ocean` significa que el resultado de la misión aterrizó con éxito en una región específica del océano, mientras que `False Ocean` significa que el resultado de la misión aterrizó sin éxito en una región específica del océano. `True RTLS` significa que el resultado de la misión aterrizó con éxito en una plataforma de tierra. `False RTLS` significa que el resultado de la misión aterrizó sin éxito en una plataforma de tierra. `True ASDS` significa que el resultado de la misión aterrizó con éxito en un barco no tripulado. `False ASDS` significa que el resultado de la misión aterrizó sin éxito en un barco no tripulado. `None ASDS` y `None None` representan un aterrizaje fallido.

```python
for i,outcome in enumerate(landing_outcomes.keys()):
    print(i,outcome)
```

Creamos un conjunto de resultados donde la segunda etapa no tuvo éxito:

```python
bad_outcomes=set(landing_outcomes.keys()[[1,3,5,6,7]])
bad_outcomes
```

## TAREA 4:
Crear una etiqueta de resultado de aterrizaje a partir de la columna Resultado

Usando Resultado, cree una lista donde el elemento sea cero si la fila correspondiente en Resultado está en el conjunto `bad_outcome`; de ​​lo contrario, es uno. Luego asígnelo a la variable `landing_class`:

```python
# landing_class = 0 si el resultado es malo
# landing_class = 1 en caso contrario

def onehot(item):
    if item in bad_outcomes:
        return 0
    else:
        return 1
landing_class = df["Outcome"].apply(onehot)
landing_class
```

Esta variable representará la variable de clasificación que representa el resultado de cada lanzamiento. Si el valor es cero, la primera etapa no aterrizó con éxito; uno significa que la primera etapa aterrizó con éxito

```python
df['Class']=landing_class
df[['Class']].head(8)
```
